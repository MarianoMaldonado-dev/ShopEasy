```python
# Paso 1: Importación de librerías | pandas, seaborn, matplotlib |
# Paso 2: Carga de datos | Lectura del archivo clientes.csv |
# Paso 3: Análisis descriptivo | Estadísticas básicas y conteo por género |
# Paso 4: Visualización | Gráfico de barras y gráfico de dispersión |
# Paso 5: Interpretación | Reflexión sobre los resultados obtenidos |

# ---

# El siguiente es el código Python completo.

# Paso 1: Importar librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Paso 2: Cargar los datos
# Para ejecutar este código, asegúrate de tener un archivo llamado 'clientes.csv'
# subido a tu entorno de Colab. Puedes subirlo usando el ícono de carpeta en el panel izquierdo.
# Luego, descomenta la siguiente línea:
# df = pd.read_csv('clientes.csv')

# --- IMPORTANTE: Si no tienes el archivo 'clientes.csv', puedes usar este DataFrame de ejemplo ---
# Si descomentas la línea anterior, elimina o comenta las siguientes líneas para usar tus propios datos.
data = {'Genero': ['Femenino', 'Masculino', 'Femenino', 'Masculino', 'Femenino'],
        'Satisfaccion': [4, 3, 5, 2, 4],
        'Gasto_Mensual': [50, 30, 70, 20, 60]}
df = pd.DataFrame(data)
# --- FIN DEL EJEMPLO DE DATOS ---

print("Primeras filas del dataset:")
print(df.head())

# Paso 3: Análisis descriptivo
print("\n--- Descripción estadística ---")
print(df.describe())
print("\n--- Conteo por género ---")
print(df['Genero'].value_counts())

# Paso 4.1: Gráfico de barras
sns.countplot(x='Genero', data=df)
plt.title("Cantidad de clientes por género")
plt.show()

# Paso 4.2: Gráfico de dispersión
sns.scatterplot(x='Satisfaccion', y='Gasto_Mensual', hue='Genero', data=df)
plt.title("Relación entre Satisfacción y Gasto Mensual")
plt.show()
```
